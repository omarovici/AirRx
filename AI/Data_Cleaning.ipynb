```python
path = "NLDAS_FORA0125_H.A20250928.1200.020.nc"
ds =xr.open_dataset(path)
```

```python
df = ds.to_dataframe().reset_index()
df

```

```python
df.isnull().sum()
```

```python
df = df.replace(-9999, np.nan)

```

```python
df = df.groupby(["lat","lon"]).apply(lambda g: g.interpolate(method="linear"))
```

```python
df = df.ffill().bfill()
```
